<p>2. Haced una exploración básica cada uno de los DataFrames, esta exploración debe incluir:</p>
<ul>
<li>¿Cuántas filas y columnas tiene el DataFrame?</li>
<li>Mostrad las primeras cinco filas y las últimas cinco filas del DataFrame.</li>
<li>Mostrad 2 filas al azar del DataFrame.</li>
<li>¿Cuál es el nombre de las columnas del DataFrame?</li>
<li>¿Cuáles son los principales estádisticos del DataFrame? Mostrad por separado las variables categóricas y las variables numéricas.</li>
<li>¿Cuál es el porcentaje de valores nulos en cada columna?</li>
<li>¿Cuántos valores únicos tenemos para cada una de las columnas?</li>
<li>De aquellas columnas que tengan menos de 6 elementos únicos mostrad cada uno de sus valores únicos.</li>
<li>📌 NOTA Fijaos que vamos a hacer lo mismo para cada archivo que tenemos. Una buena práctica sería crearnos una función que apliquemos a cada conjunto de datos. Nuestra función recibirá un parámetro, que será el dataframe entero </li>
<li>📌 NOTA Nos estamos convirtiendo en analistas de datos y no nos debe valer solo con escribir el código, os recomendamos que reviseis bien los resultados y empezais a extraer algunas conclusiones. ¿Hay datos raros en nuestros conjuntos de datos? ¿Tenemos nulos?, etc.</li>
</ul>

```python
def dataframe_exploracion (dataframe):
    print(f'El dataframe tiene forma {dataframe.shape}')
    print(f'Las primeras cinco filas del DataFrame son: \n {dataframe.head()}')
    print(f"Las cinco últimas filas del Dataframe son: \n {dataframe.tail()}")
    print(f"El nombre de las columnas del Dataframe es", dataframe.columns)
    print(f'Los principales estadísticos del Dataframe son', dataframe.describe())
    print(f'El porcentaje de valores nulos es igual: {dataframe.isnull().sum()/100} %')
    print(f'Los valores únicos para cada columna son:')
    for col in dataframe.columns:
        print(f'El número de valores únicos de la columna {col} es {len(dataframe[col].unique())}')
    for col in dataframe.columns:
        if len(dataframe[col].unique()) > 6: 
            print(f"Los valores únicos de {col} son: {dataframe[col].unique()}")
```

```python
dataframe_exploracion(dataframe_agentes)
```

```python
dataframe_exploracion(dataframe_2016)
```

```python
dataframe_exploracion(dataframe_empleados)
```
